```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd

d = 1.21 #kg/m^3 density of air
R = 1 #m radius of ball
C_d = 0.24 #coefficient of drag for tennis ball
m = 0.43 #kg mass of ball
k = np.pi*d*R*R/(2*m)
g = 9.81 #m/s^2 
omega = 100 #rad/sec angular velocity of ball
k_1 = 0.0025
k_2 = 0.0228


v = lambda a, b: np.sqrt(a**2+b**2)
dvy = lambda vvx, vvy: (k_1*omega*vvx-k_2*C_d*v(vvx, vvy)*vvy-g*m)/m
dvx = lambda vvx, vvy: (-1*k_1*omega*vvy-k_2*C_d*v(vvx, vvy)*vvx)/m



dur = 2
dt = 0.001
n_steps = int(dur/dt)
dur = n_steps*dt
t = 0

x = np.zeros(n_steps)
y = np.zeros(n_steps)
vx = np.zeros(n_steps)
vy = np.zeros(n_steps)

vx[0] = 0
vy[0] = 3

for i in range(n_steps-1):
  vx[i+1] = vx[i]+dvx(vx[i], vy[i])*dt
  vy[i+1] = vy[i]+dvy(vx[i], vy[i])*dt
  x[i+1] = x[i]+vx[i+1]*dt
  y[i+1] = y[i]+vy[i+1]*dt

df = pd.DataFrame()
df["x (m)"] = x.tolist()
df["y (m)"] = y.tolist()
fig = px.line(df, x="x (m)", y="y (m)", title='Life expectancy in Canada')
fig.show()
```
